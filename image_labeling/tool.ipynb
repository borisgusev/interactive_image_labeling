```python
from labelingtool import LabelingTool
import pandas as pd
from tifffile import imread
import matplotlib.pyplot as plt

database = "../training_database_classifier.csv"
column_raw = "raw"
column_mask = "sobel"

classes = ["worm", "egg", "error"]
datab_df = pd.read_csv(database)
raw_images = list(datab_df[column_raw])
sobel_images = list(datab_df[column_mask])
```

```python
tool = LabelingTool(classes=classes, images=sobel_images, second_images=raw_images)
```

```python
tool.start()
```

```python
print(len(tool.labels))
```

```python
pd.DataFrame(tool.labels.items(), columns=["file_path", "label"]).to_csv("labels.csv", index=False)
```
